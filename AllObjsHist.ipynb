```python
import matplotlib.pyplot as plt
import json
```

```python
with open('roman_data/annotations/all_metadata.json', 'r') as f:
    all_data = json.load(f)
```

```python
len(all_data)
```

```python
# hist of objects vs num of images in test set
objects = []
for d in all_data:
    objects.append(len(d['annotations']))
plt.hist(objects, bins=40, color='orange', alpha=0.7)
plt.xlabel('Number of Objects')
plt.ylabel('Number of Images')
plt.title('All Objs')
print(sum(objects))
```

```python
with open('roman_data/annotations-combined/train_roman.json', 'r') as f:
    alld = json.load(f)

with open('roman_data/annotations-combined/test_roman.json', 'r') as f:
    test_data = json.load(f)

alld.extend(test_data)

with open('roman_data/annotations-combined/val_roman.json', 'r') as f:
    val_data = json.load(f)

alld.extend(val_data)
```

```python
len(alld)
```

```python
# hist of objects vs num of images in test set
objects_c = []
for d in alld:
    objects_c.append(len(d['annotations']))
plt.hist(objects_c, bins=40, color='orange', alpha=0.7)
plt.xlabel('Number of Objects')
plt.ylabel('Number of Images')
plt.title('All Objs in LSST Set')
print(sum(objects_c))
```

```python
combined = False
run_dir = './roman_runs/run9-swin'
# run_dir = './roman_runs/run20-combined-swin'
if 'combined' in run_dir:
    combined = True

if combined:
    with open('roman_data/annotations-combined/train_roman.json', 'r') as f:
        train_data = json.load(f)
else:
    with open('roman_data/annotations/train_roman.json', 'r') as f:
        train_data = json.load(f) 
        
if combined:
    with open('roman_data/annotations-combined/val_roman.json', 'r') as f:
        val_data = json.load(f)
else:
    with open('roman_data/annotations/val_roman.json', 'r') as f:
        val_data = json.load(f) 
        
if combined:
    with open('roman_data/annotations-combined/test_roman.json', 'r') as f:
        data = json.load(f)
else:
    with open('roman_data/annotations/test_roman.json', 'r') as f:
        data = json.load(f)  
        
```

```python
# hist of objects vs num of images in train set
train_objects = []
test_objs = []
val_objs = []
for d in train_data:
    train_objects.append(len(d['annotations']))
for d in data:
    test_objs.append(len(d['annotations']))
for d in val_data:
    val_objs.append(len(d['annotations']))
plt.hist(train_objects, bins=40, alpha=0.8, label='Training Set')
plt.hist(val_objs, bins=40, alpha=0.7, label='Validation Set')
plt.hist(test_objs, bins=40, alpha=0.8, label='Test Set')
plt.legend()
plt.xlabel('Number of Objects')
plt.ylabel('Number of Images')
plt.title('Distribution of Data - Roman')
print(sum(train_objects))
print(sum(val_objs))
print(sum(test_objs))
```
